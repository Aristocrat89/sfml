```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
#Зависимость ROC_AUC от параметра Max_depth для Decision Tree Classifier 
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score
x = range(1, 20)   #Лист с набором значений для параметра max_depth
y = []    #Лист, в который будут записываться результаты roc_auc_score для каждого значения max_depth
for i in x:
    dtc = DecisionTreeClassifier(max_depth=i)
    dtc.fit(X_train, y_train)
    y.append(roc_auc_score(y_test, dtc.predict_proba(X_test)[:,1]))
plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.ylabel('Roc_auc')
plt.xlabel('Max_depth')
plt.grid(True)
plt.title('Decision Tree Classifier')
plt.xlim((0.0, 20))
plt.ylim((0.6, 0.9))
```

```python
#Зависимость ROC_AUC от параметра N_neighbors для K_Neighbors Classifier
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score
x = range(1, 10)    #Лист с набором значений для параметра n_neighbors
y = []     #Лист, в который будут записываться результаты roc_auc_score для каждого значения n_neighbors
for i in x:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y.append(roc_auc_score(y_test, knn.predict_proba(X_test)[:,1]))
plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.ylabel('Roc_auc')
plt.xlabel('N_neighbors')
plt.grid(True)
plt.title('K_Neighbors Classifier')
plt.xlim((0.0, 10))
plt.ylim((0.0, 0.7))
```

```python
#Зависимость ROC_AUC от параметра С для Logistic Regression
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score
x = range(1, 20)    #Лист с набором значений для параметра С
y = []     #Лист, в который будут записываться результаты roc_auc_score для каждого значения C
for i in x:
    lr = LogisticRegression(C=i)
    lr.fit(X_train, y_train)
    y.append(roc_auc_score(y_test, lr.predict_proba(X_test)[:,1]))
plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.ylabel('Roc_auc')
plt.xlabel('C')
plt.grid(True)
plt.title('Logistic Regression')
plt.xlim((0.0, 20))
plt.ylim((0.5, 0.7))
```

```python
#Зависимость ROC_AUC от параметров C и Penalty для Logistic Regression
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score
x = range(1, 10)     #Лист с набором значений для параметра penalty
y1 = []      #Лист, в который будут записываться результаты roc_auc_score для каждого значения C при penalty='l1'
y2 = []      #Лист, в который будут записываться результаты roc_auc_score для каждого значения C при penalty='l2'
for i in x:
    lr = LogisticRegression(penalty='l1', C=i)
    lr.fit(X_train, y_train)
    y1.append(roc_auc_score(y_test, lr.predict_proba(X_test)[:,1]))
    lr = LogisticRegression(penalty='l2', C=i)
    lr.fit(X_train, y_train)
    y2.append(roc_auc_score(y_test, lr.predict_proba(X_test)[:,1]))
plt.figure(figsize=(10, 10))
plt.plot(x, y1)
plt.plot(x, y2)
plt.ylabel('Roc_auc')
plt.xlabel('C')
plt.grid(True)
plt.title('Logistic Regression')
plt.xlim((0.0, 10))
plt.ylim((0.0, 1))
```
